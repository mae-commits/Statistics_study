## Hierarchic clustering

```python
# Import libraries to draw figures and graphs.

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
# Import libraries in relation to manipulate the numerical calculation and data frame.

import numpy as np 
import pandas as pd
from pandas import plotting
```

```python
# Import libraries to offer access to resources.
import urllib.request
```

```python
# Decide resources on the web.

url = 'https://raw.githubusercontent.com/maskot1977/ipython_notebook/master/toydata/sake_dataJ.txt'

# Rename the resources name.

urllib.request.urlretrieve(url, 'sake_data1.txt')
```

```python
# Read the data.

df = pd.read_csv('sake_data1.txt', sep='\t', index_col=0)
df
```

### Normalize the matrix.

```python
# Normalize the matrix.

dfs = df.apply(lambda x: (x-x.mean())/x.std(), axis=0).fillna(0)
dfs
```

### Execute the clustering.

```python
# Choose the metric and method.

from scipy.cluster.hierarchy import linkage, dendrogram
result1 = linkage(dfs.iloc[:, :], 
                  # metric = 'braycurtis', 
                  # metric = 'canberra', 
                  # metric = 'chebyshev', 
                  # metric = 'cityblock', 
                  metric = 'correlation', 
                  # metric = 'cosine', 
                  # metric = 'euclidean', 
                  # metric = 'hamming', 
                  # metric = 'jaccard', 
                  # method= 'single')
                  method = 'average')
                  # method= 'complete')
                  # method='weighted')
                  
plt.figure(figsize=(8, 8))
dendrogram(result1, orientation='right', labels=list(df.index), color_threshold=0.8)
plt.title("Dedrogram")
plt.xlabel("Threshold")
plt.grid()
plt.show()
```

```python
# Choose the metric and method.
# Reverse the lines and rows.

from scipy.cluster.hierarchy import linkage, dendrogram
result1 = linkage(dfs.T.iloc[:, :], 
                  # metric = 'braycurtis', 
                  # metric = 'canberra', 
                  # metric = 'chebyshev', 
                  # metric = 'cityblock', 
                  metric = 'correlation', 
                  # metric = 'cosine', 
                  # metric = 'euclidean', 
                  # metric = 'hamming', 
                  # metric = 'jaccard', 
                  # method= 'single')
                  method = 'average')
                  # method= 'complete')
                  # method='weighted')
                  
plt.figure(figsize=(8, 8))
dendrogram(result1, orientation='right', labels=list(df.columns), color_threshold=0.8)
plt.title("Dedrogram")
plt.xlabel("Threshold")
plt.grid()
plt.show()
```
